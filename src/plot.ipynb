```python
from adjustText import adjust_text
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data from the image with model names
data = {
    'Model': [
    'Linear Regression',
    'Lasso Regression',
    'Ridge Regression',
    'Elastic Net Regression',
    'Beta Regression',
    'XGBoost',
    'Random Forest',
    'LightGBM',
    'Catboost',
    'Cubist',
    'SVM',
    'MLP',
    'Resnet',
    'FT Transformer',
    'TabPFN',
    'GANDALF',
    'GANDALF_bagging',
    'Exact Gaussian Process',
    'Power Expectation Propagation',
    'Elastic Net Meta Learner'
],
    'RMSE':[
    0.286, 0.291, 0.276, 0.278, 0.293,
    0.214, 0.221, 0.225, 0.223, 0.219,
    0.255, 0.259, 0.242, 0.250, 0.218,
    0.248, 0.232, 0.219, 0.222, 0.212
]
,
    'Training_Time': [
    28.077, 1.845, 1.468, 235.082, 60,
    105.197, 13.796, 7.054, 17.608, 22.9,
    0.141, 551.4, 24.2, 515.221, 2,
    480, 1380, 63, 3179.523, 0.001
]

}
```

```python
# Create the plot
plt.figure(figsize=(12, 8))
plt.scatter(data['Training_Time'], data['RMSE'], color='blue', alpha=0.6, s=100)

# Labels and title
plt.xlabel('Training Time (seconds)', fontsize=12)
plt.ylabel('Out-of-sample RMSE', fontsize=12)
plt.title('Model Performance: RMSE vs Training Time', fontsize=14, pad=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xscale('log')

# Add text annotations using adjustText
texts = []
for x, y, model in zip(data['Training_Time'], data['RMSE'], data['Model']):
    texts.append(plt.text(x, y, model, fontsize=10))

adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

plt.tight_layout()
plt.show()
```
